# Step 0 Implementation

```python
import numpy as np 
import pandas as pd
import re

energy = pd.read_excel("Energy Indicators.xls" ,index_col = None ,header = None)
energy = energy[18:245].reset_index()
energy.drop([0,1], axis = 1, inplace = True)
energy.set_index('index', inplace = True)

energy.rename(columns = {2:'Country' ,3:'Energy Supply' ,4:'Energy Supply per Capita' ,5:'% Renewable' }, inplace = True)

energy.replace({"..." : np.nan} , inplace = True)

energy['Energy Supply'] *= 1000000

energy["Country"] = energy['Country'].str.replace('\d+', "") 


country_names = list(energy["Country"]) 
for i in range(len(country_names)): 
    country_names[i] = re.sub(r"\([^()]*\)", "", country_names[i]).strip()
energy["Country"] = country_names
 
energy['Country'].replace({"Republic of Korea": "South Korea",
"United States of America": "United States",
"United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
"China, Hong Kong Special Administrative Region": "Hong Kong"}, inplace = True)

energy.head(60)
```

# step 1 Implementation

```python
GDP = pd.read_csv('world_bank.csv',index_col = None ,header = 4)
GDP['Country Name'].replace({"Korea, Rep.": "South Korea", 
"Iran, Islamic Rep.": "Iran",
"Hong Kong SAR, China": "Hong Kong"}, inplace=True)
GDP.rename(columns = {"Country Name": "Country"}, inplace = True)
GDP.head(60)
```

# Step 2 Implementation

```python
ScimEn = pd.read_excel("scimagojr-3.xlsx")
ScimEn
```

# Step 3 Implementation

```python
ScimEn_energy_join = pd.merge(ScimEn, energy [[ 'Country','Energy Supply', 'Energy Supply per Capita', '% Renewable']], on = 'Country')
ScimEn_energy_GDP_join = pd.merge(ScimEn_energy_join,GDP [['Country','2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']], on = 'Country')
ScimEn_energy_GDP_join.set_index('Country', inplace = True)
Full_Join = ScimEn_energy_GDP_join.drop(ScimEn_energy_GDP_join.index[15:])
Full_Join
```

# step 8 Implementation

```python
Full_Join["Ratio Citations"] = Full_Join["Self-citations"] / Full_Join["Citations"]
print("The Maximum Ratio is :", Full_Join["Ratio Citations"].max())
print("The Maximum Ratio Country is :", Full_Join["Ratio Citations"].idxmax())
print("Result : ", tuple([Full_Join["Ratio Citations"].idxmax(),Full_Join["Ratio Citations"].max() ]))
```

# step 9 Implementation

```python
MedianCheck = []
for i in Full_Join['% Renewable']:     
    if i > Full_Join['% Renewable'].median():
        MedianCheck.append('1')
    else: MedianCheck.append('0')
Full_Join['median check'] = MedianCheck
Full_Join
```

# step 10 Implementation

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 'Japan':'Asia', 'United Kingdom':'Europe', 'Russian Federation':'Europe', 
                  'Canada':'North America', 'Germany':'Europe','India':'Asia','France':'Europe','South Korea':'Asia', 'Italy':'Europe', 
                  'Spain':'Europe', 'Iran':'Asia','Australia':'Australia', 'Brazil':'South America'}

copy_full_join = Full_Join.copy()
copy_full_join.reset_index(level = 0 , inplace = True)

copy_full_join['Continent'] = [(list(ContinentDict.values())[list(ContinentDict.keys()).index(item)]) if item in ContinentDict.keys() else item for item in copy_full_join['Country'] ]

copy_full_join['estimated population'] = copy_full_join['Energy Supply']/copy_full_join['Energy Supply per Capita']

copy_full_join.drop(['Rank','Country','Energy Supply','Energy Supply per Capita','H index','Documents','Citable documents','Citations','Self-citations','Citations per document','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','Ratio Citations','median check', '% Renewable'], axis = 1, inplace = True)
new_data = pd.DataFrame(columns = ['size','sum','mean','std'])
new_data['size'] = copy_full_join.groupby('Continent').size()
new_data['sum'] = copy_full_join.groupby('Continent').sum()
new_data['mean'] = copy_full_join.groupby('Continent').mean()
new_data['std'] = copy_full_join.groupby('Continent').std()
new_data
```
